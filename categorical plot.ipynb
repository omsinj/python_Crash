```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
sns.barplot(tips, x='sex', y = 'total_bill', estimator = np.mean)
```

```python
sns.countplot(data = tips, x = 'sex')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data = tips, hue= 'smoker', split = True)
```

```python
sns.stripplot(x='day', y= 'total_bill', data= tips, jitter = True, hue = 'smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data = tips, hue= 'smoker', split = True)
sns.swarmplot(x='day', y='total_bill', data = tips, native_scale=True, color = 'black')
```

```python
sns.catplot (data=tips, x = 'day', y = 'total_bill', kind = 'box')
sns.catplot (data=tips, x = 'day', y = 'total_bill', kind = 'bar')
```
