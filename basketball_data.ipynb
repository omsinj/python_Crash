```python
def myplot(playerlist):
    for name in playerlist:
        plt.plot(Salary[Pdict[name]], c = "red", ls= "--", marker="s", ms=7, label = name)
        plt.legend(loc = 'upper left', bbox_to_anchor=(0,1))
        plt.xticks(list(range(0,10)), Seasons, rotation= 'vertical')
        plt.show()
```

```python
myplot(["LeBronJames", "DerrickRose"])
```

```python
def myplot(playerlist):
    Col = {"LeBronJames": "Red", "DerrickRose":"Blue"}
    for name in playerlist:
        plt.plot(Salary[Pdict[name]], c = Col[name], ls= "--", marker="s", ms=7, label = name)
        plt.legend(loc = 'upper left', bbox_to_anchor=(0,1))
        plt.xticks(list(range(0,10)), Seasons, rotation= 'vertical')
        plt.show()
```

```python
def myplot(playerlist):
    Col = {"LeBronJames": "Red", "DerrickRose": "Blue"}
    Mrk = {"LeBronJames": "s", "DerrickRose": "o"}
    
    for name in playerlist:
        plt.plot(Salary[Pdict[name]], c=Col[name], ls="--", marker=Mrk[name], ms=7, label=name)
    
    plt.legend(loc='upper left', bbox_to_anchor=(0,1))
    plt.xticks(list(range(0,10)), Seasons, rotation= 'vertical')
    plt.show()
myplot(["LeBronJames", "DerrickRose"])
```
