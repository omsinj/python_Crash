```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

```python
sns.lmplot(data = tips, x='total_bill', y = 'tip', hue = 'sex', markers = ['o', 'v'], 
          scatter_kws= {'s': 100})
```

```python
sns.lmplot(data = tips, x='total_bill', y = 'tip', col = 'sex', row = 'time')
```

```python
sns.lmplot(data = tips, x='total_bill', y = 'tip',hue = 'sex', col = 'day', row = 'time')
```
