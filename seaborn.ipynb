```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
```

```python
stats = pd.read_csv("demo.csv")
```

```python
stats.columns
```

```python
stats.columns = ["CountryName", "CountryCode", "BirthRate", "InternetUsers", "IncomeGroup" ]
```

```python
vis1 = sns.displot(stats["InternetUsers"])
```

```python
vis2 = sns.displot(stats["InternetUsers"], bins=30)
```

```python
vis3 =sns.displot(stats["InternetUsers"], bins=30, color = 'Red')
```

```python
vis4 = sns.boxplot(stats, x="IncomeGroup", y="BirthRate")
```

https://seaborn.pydata.org/examples/index.html

```python
import matplotlib as mpl
sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(7, 5))
sns.despine(f)

sns.histplot(
    stats,
    x="BirthRate", hue="BirthRate",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5,
    log_scale=True,
)
ax.xaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
ax.set_xticks([500, 1000, 2000, 5000, 10000])

plt.show()
```

```python
vis5 = sns.lmplot(data = stats, x='InternetUsers', y='BirthRate', fit_reg = False)
```

```python
vis6 = sns.lmplot(data = stats, x='InternetUsers', y='BirthRate', fit_reg = False, hue='IncomeGroup')
```

```python
vis7 = sns.lmplot(data = stats, x='InternetUsers', y='BirthRate', fit_reg = False, hue='IncomeGroup', aspect = 2)
```

```python
vis7 = sns.lmplot(data = stats, x='InternetUsers', y='BirthRate', fit_reg = False, hue='IncomeGroup', aspect = 2, height=10)
```

```python
vis8 = sns.lmplot(data = stats, x='InternetUsers', y='BirthRate', fit_reg = False, hue='IncomeGroup', aspect = 2, height=10, markers = 'o')
```
